<h1>This one is for Kinesis Streams producer</h1>
<h3>Kinesis Firehose Features</h3>
<ul>
  <li>Real-time data processing with low latency capabilty</li>
  <li>Not Completely managed service if you select <b>'Provisioned'</b>, required little configuration on shard level</li>
  <li>Data can be stored as per retention policy</li>
</ul>

```python
#This one is for Kinesis Streams producer#
#pip install Faker
#pip install boto3

from faker import Faker
import random as rd
import json
import boto3

access_key = "type your access key"
secret_key = "type your secret key"
region = "ap-south-1"
session = boto3.Session(aws_access_key_id=access_key, aws_secret_access_key=secret_key, region_name=region)

client = session.client('kinesis')
kinesis_stream = "demo-kinesis-stream-1"
sample = Faker()  

def getData(id, name, country, source):
  data = {}
  data["id"] = id
  data["name"] = name
  data["country"] = country
  data["source"] = source
  return data

temp = 0
while temp<50:
  if temp<=25:
    record = json.dumps(getData(str(rd.randint(1, 100)),sample.name(),sample.country(),'local_system'))
    client.put_record(StreamName=kinesis_stream, Data=record, PartitionKey="1")
  else:
    record = json.dumps(getData(str(rd.randint(1, 100)),sample.name(),sample.country(),'other'))
    client.put_record(StreamName=kinesis_stream, Data=record, PartitionKey="1")
  print(record)
  temp = temp+1
```

<h1>This one is for Kinesis Firehose Direct Put producer</h1>
<h3>Kinesis Firehose Features</h3>
<ul>
  <li>Near real-time data processing (depends on buffer size or minimum buffer time of 60 seconds)</li>
  <li>Mostly used for data transfer</li>
  <li>Completely managed service</li>
  <li>No data storage capacity</li>
</ul>

```python
#This one is for Kinesis Firehose Direct Put producer#
#pip install Faker
#pip install boto3

from faker import Faker
import random as rd
import json
import boto3

access_key = "type your access key"
secret_key = "type your secret key"
region = "ap-south-1"
session = boto3.Session(aws_access_key_id=access_key, aws_secret_access_key=secret_key, region_name=region)

client = session.client('firehose')
kinesis_stream = "demo-firehose-stream-1"
sample = Faker()  

def getData(id, name, country, source):
  data = {}
  data["id"] = id
  data["name"] = name
  data["country"] = country
  data["source"] = source
  return data

temp = 1
while temp<=100:
  if temp<=50:
    record = json.dumps(getData(rd.randint(1, 100),sample.name(),sample.country(),'databricks'))
    client.put_record(DeliveryStreamName=kinesis_stream,
    Record={
        'Data': record
    })
  else:
    record = json.dumps(getData(rd.randint(1, 100),sample.name(),sample.country(),'other'))
    client.put_record(DeliveryStreamName=kinesis_stream,
    Record={
        'Data': record
    })
  #print(record)
  temp = temp+1
```
